# Handling All missing/invalid value

## Price Per Unit

```python
# here i fill all UNKNOWN, ERROR, NaN with 0
# i will expect the price with its type of item , location

data["Price Per Unit"].replace(['UNKNOWN', 'ERROR'], 0 , inplace=True)
data["Price Per Unit"].fillna(0, inplace=True)
data["Price Per Unit"]=data["Price Per Unit"].astype(float)
data["Price Per Unit"].unique()
```

## Quantity

```python
# here i fill all UNKNOWN, ERROR, NaN with 0
# i will expect the Quantity with its type of total price



data["Quantity"].replace(['UNKNOWN', 'ERROR'], 0 , inplace=True)
data["Quantity"].fillna(0, inplace=True)
data["Quantity"]=data["Quantity"].astype(int)
data["Quantity"].unique()
```

## Location

```python
# here i fill all ERROR, NaN with UNKNOWN

data["Location"].replace(['ERROR'], 'UNKNOWN' , inplace=True)
data["Location"].fillna('UNKNOWN', inplace=True)
data["Location"].unique()
```

## Payment Method

```python
# here i fill all ERROR, NaN with UNKNOWN

data["Payment Method"].replace(['ERROR'], 'UNKNOWN' , inplace=True)
data["Payment Method"].fillna('UNKNOWN', inplace=True)
data["Payment Method"].unique()
```

## Item

```python
# here i fill all ERROR, NaN with UNKNOWN

data["Item"].replace(['ERROR'], 'UNKNOWN' , inplace=True)
data["Item"].fillna('UNKNOWN', inplace=True)
data["Item"].unique()
```

## Total Spent

```python
# here i fill all UNKNOWN, ERROR, NaN with 0

data["Total Spent"].replace(['UNKNOWN', 'ERROR'], 0 , inplace=True)
data["Total Spent"].fillna(0, inplace=True)
data["Total Spent"]=data["Total Spent"].astype(float)
data["Total Spent"].unique()
```

## Transaction Date

```python
# change to Date Type
data['Transaction Date'] = pd.to_datetime(data['Transaction Date'], errors='coerce')
# Forward Fill
data['Transaction Date'] = data['Transaction Date'].ffill()
```

## Make filtering to know the price of unit

```python
# I use this filter to know the price of each item
# location not impact on price

result = data.groupby('Item').agg({
    'Price Per Unit': 'unique',
    'Location': 'unique'
}).reset_index()
result
```

# تم عمل استبدال لكل القيم غير الصالحه و القيم الفارغه

```python
data.info()
```

## convert price_to_item

```python
# I replace the missing values with its actual item based on its price
price_to_item_map = {
    3.0: 'Cake',
    2.0: 'Coffee',
    1.0: 'Cookie',
    5.0: 'Salad',
    4.0: 'Sandwich',
    1.5: 'Tea'
}

mask_item = (data['Item'] == 'UNKNOWN')

data.loc[mask_item, 'Item'] = data.loc[mask_item, 'Price Per Unit'].map(price_to_item_map)
data["Item"].unique()
```

## convert Item to price

```python
# I have a price of items
# so i fill the null in item column with its price

Item_price = {
  'Cake': 3.0,
  'Coffee': 2.0,
  'Cookie': 1.0,
  'Juice': 3.0,
  'Salad': 5.0,
  'Sandwich': 4.0,
  'Smoothie': 4.0,
  'Tea': 1.5
}

mask = (data['Price Per Unit'] == 0)

data.loc[mask, 'Price Per Unit'] = data.loc[mask, 'Item'].map(Item_price)

data['Price Per Unit'].unique()
```

```python
data = data.dropna(subset=['Item'])
```

```python
data.isna().sum()
```

# تم عمل استبدال للقيم غير الصالحه ب القيم الحقيقه عن طريق قسم القيمه الكليه علي سعر الوحده الواحده

```python

mask = (data['Quantity'] == 0)

data.loc[mask, 'Quantity'] = data.loc[mask, 'Total Spent'] / data.loc[mask, 'Price Per Unit']
```

# تم حذف القيم التي بها قيم غير صالحه و عددهم  هو 20 قيمه حتي تكون الداتا نظيفه 100 %

```python
# drop to 20 row with zero value
data = data[data['Quantity'] != 0]

data = data.reset_index(drop=True)
```

## Total spent with 0 value (was NaN)

# تم عمل حساب للقيمه الكليه عن طريق ضرب الكميه المطلوبه مع سعر الوحده الواحده

```python
# Total Spent = Quantity *Price Per Unit
# but with zero (was NaN and convert to 0 )
mask = (data['Total Spent'] == 0)

data.loc[mask, 'Total Spent'] = data.loc[mask, 'Quantity'] * data.loc[mask, 'Price Per Unit']
data["Total Spent"].unique()
```
